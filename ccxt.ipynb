```python
@dataclass
class PositionData:
    accountValue: float
    positionMargin: float

def get_positions_data() -> PositionData :
    full_data = dex.fetch_balance()
    print(full_data)
    return PositionData(
        accountValue=float(full_data['crossMarginSummary']['accountValue']),
        positionMargin=float(full_data['crossMarginSummary']['totalMarginUsed']))

```

```python
def updateOrder(incAmount):
    orders = dex.fetch_open_orders(symbol)

    for o in orders:
        print("Order", o['id'], "Price", o['price'], "Amount", o['amount'])
        print("Updating order", o['id'])
        dex.cancel_order(o['id'], symbol=symbol)
        new_price = o['price'] + incAmount
        new_order = dex.create_order(symbol, buy, 'limit', o['amount'], new_price)
        print("New order created", new_order)
        print (' -- - - - - ')

updateOrder(-1400)
```
